```python
import pandas as pd
import numpy as np
```

```python
dirs = ['R1398','R1399','R1400','R1401','U1398','U1399','U1400','U1401']
for dir in dirs :
    df =pd.read_csv(f'{dir}/{dir}P3S14.csv')
    df['value']=df['value'].apply(lambda x : str(x).lstrip('0'))
    df.loc[(df.value.isna())|(df.value=='nan') | (df.value==''),'value']=np.nan
    df.replace(r'^\s*$| ', np.nan, regex=True,inplace=True)
    print(df.loc[df.value.isna(),'Address'].count())
    df['value'] = pd.to_numeric(df['value'])
    df['value'] = df['value'].fillna(df.groupby('code')['value'].transform('median'))
    print('addresses still have nan value: ',df[df.value.isna()].code.count())
    print(df[df.isna().any(axis=1)])
    df = df[df['value'].notna()]
    print('nan value after cleaning: ',df[df.value.isna()].Address.count())
    df.to_csv(f'{dir}/{dir}P3S14.csv',index=False)
    print("----------------------------------")

```
